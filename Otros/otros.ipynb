```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
import numpy as np
```

```python
#df_sentences = data.iloc[:, 50]

sentences_list = data['motivo_name'].tolist()
sentences_list  = sentences_list[:50]
```

```python
sentences_list
```

```python
sentences = sentences_list
```

```python
sentences = ["carrera 79 # 19 87", 
       "cra 79 número 19 - 87", 
       "kra 79 # 19 87",
       "carrera 79 # 19 87 la felcidad",
       "cra 79 número 19 - 87, conjunto la fuente int 4 apt 1004"]
```

```python
# Tokenization of each document
tokenized_sent = []
for s in sentences:
    tokenized_sent.append(word_tokenize(s.lower()))
tokenized_sent
```

```python
def cosine(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
```

**Doc2Vec**

```python
# import
#!pip install gensim
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
tagged_data = [TaggedDocument(d, [i]) for i, d in enumerate(tokenized_sent)]
tagged_data
```

```python
## Train doc2vec model
model = Doc2Vec(tagged_data, vector_size = 20, window = 2, min_count = 1, epochs = 100)

'''
vector_size = Dimensionality of the feature vectors.
window = The maximum distance between the current and predicted word within a sentence.
min_count = Ignores all words with total frequency lower than this.
alpha = The initial learning rate.
'''

## Print model vocabulary
model.wv.vocab
```

```python
test_doc = word_tokenize("caye 25 # 1 87".lower())
test_doc_vector = model.infer_vector(test_doc)
model.docvecs.most_similar(positive = [test_doc_vector])
```

**SentenceBERT**

```python
#!pip install sentence_transformers
from sentence_transformers import SentenceTransformer
sbert_model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
sentence_embeddings = sbert_model.encode(sentences)
```

```python
query = "CALLE 25 # 1 87"
query_vec = sbert_model .encode([query])[0]
```

```python
from doqu import InferSent
import torch

V = 2
MODEL_PATH = 'encoder/infersent%s.pkl' % V
params_model = {'bsize': 64, 'word_emb_dim': 300, 'enc_lstm_dim': 2048,
                'pool_type': 'max', 'dpout_model': 0.0, 'version': V}
model = InferSent(params_model)
model.load_state_dict(torch.load(MODEL_PATH))

W2V_PATH = '/content/GloVe/glove.840B.300d.txt'
model.set_w2v_path(W2V_PATH)
```

```python
model.build_vocab(sentences, tokenize=True)
```

```python
query = "CALLE 25 # 1 87"
query_vec = model.encode(query)[0]
query_vec
```

```python
similarity = []
for sent in sentences:
  sim = cosine(query_vec, model.encode([sent])[0])
  print("Sentence = ", sent, "; similarity = ", sim)
```
